### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tienda2.head()
tienda3.head()
tienda4.head()
```

#1. Análisis de facturación



```python
# Crear columna de facturación en cada tienda
tienda['facturacion'] = tienda['Precio'] * tienda['Cantidad de cuotas']

tienda2['facturacion'] = tienda2['Precio'] * tienda2['Cantidad de cuotas']
tienda3['facturacion'] = tienda3['Precio'] * tienda3['Cantidad de cuotas']
tienda4['facturacion'] = tienda4['Precio'] * tienda4['Cantidad de cuotas']

# Sumar las facturaciones totales de cada tienda
facturacion_total_tienda1 = tienda['facturacion'].sum()
facturacion_total_tienda2 = tienda2['facturacion'].sum()
facturacion_total_tienda3 = tienda3['facturacion'].sum()
facturacion_total_tienda4 = tienda4['facturacion'].sum()

print(f"Tienda 1:{facturacion_total_tienda1}\nTienda 2:{facturacion_total_tienda2} \nTienda 3:{facturacion_total_tienda3} \nTienda 4:{facturacion_total_tienda4}")
```

```python
# Agrupar por vendedor y calcular la facturación
facturacion_por_vendedor = tienda.groupby('Vendedor')['facturacion'].sum().reset_index()
facturacion_por_vendedor2 = tienda2.groupby('Vendedor')['facturacion'].sum().reset_index()
facturacion_por_vendedor3 = tienda3.groupby('Vendedor')['facturacion'].sum().reset_index()
facturacion_por_vendedor4 = tienda4.groupby('Vendedor')['facturacion'].sum().reset_index()

# Combinar los resultados de las facturaciones de cada tienda
facturacion_total = pd.concat([
    facturacion_por_vendedor.set_index('Vendedor'),
    facturacion_por_vendedor2.set_index('Vendedor'),
    facturacion_por_vendedor3.set_index('Vendedor'),
    facturacion_por_vendedor4.set_index('Vendedor')
], axis=1, keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

# Mostrar la tabla combinada
display(facturacion_total)
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y calcular la facturación
facturacion_por_categoria = tienda.groupby('Categoría del Producto')['facturacion'].sum().reset_index()
facturacion_por_categoria2 = tienda2.groupby('Categoría del Producto')['facturacion'].sum().reset_index()
facturacion_por_categoria3 = tienda3.groupby('Categoría del Producto')['facturacion'].sum().reset_index()
facturacion_por_categoria4 = tienda4.groupby('Categoría del Producto')['facturacion'].sum().reset_index()


facturacion_total = pd.concat([
    facturacion_por_categoria.set_index('Categoría del Producto'),
    facturacion_por_categoria2.set_index('Categoría del Producto'),
    facturacion_por_categoria3.set_index('Categoría del Producto'),
    facturacion_por_categoria4.set_index('Categoría del Producto')
], axis=1, keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

# Mostrar la tabla combinada
display(facturacion_total)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio por tienda
calificacion_promedio_tienda = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Crear un DataFrame con las calificaciones promedio de cada tienda
calificaciones = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_promedio_tienda, calificacion_promedio_tienda2,
                              calificacion_promedio_tienda3, calificacion_promedio_tienda4]
})

# Calcular el promedio de las calificaciones promedio
promedio_total = calificaciones['Calificación Promedio'].mean()

print(f"El promedio total de las calificaciones es: {promedio_total}")
# Mostrar la tabla con las calificaciones y el promedio total
display(calificaciones)
```

# 4. Productos más y menos vendidos

```python
# Agrupar por producto y calcular la facturación
facturacion_por_producto = tienda.groupby('Producto')['facturacion'].sum().sort_values(ascending=True).reset_index()
facturacion_por_producto2 = tienda2.groupby('Producto')['facturacion'].sum().sort_values(ascending=True).reset_index()
facturacion_por_producto3 = tienda3.groupby('Producto')['facturacion'].sum().sort_values(ascending=True).reset_index()
facturacion_por_producto4 = tienda4.groupby('Producto')['facturacion'].sum().sort_values(ascending=True).reset_index()


# Combinar los resultados de las facturaciones de cada tienda
facturacion_total = pd.concat([
    facturacion_por_producto.set_index('Producto'),
    facturacion_por_producto2.set_index('Producto'),
    facturacion_por_producto3.set_index('Producto'),
    facturacion_por_producto4.set_index('Producto')
], axis=1, keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

facturacion_total = facturacion_total.reset_index()

# Mostrar la tabla combinada
display(facturacion_total)
```

```python
# Agrupar por producto y calcular la facturación
facturacion_por_producto = tienda.groupby('Producto')['facturacion'].sum().sort_values(ascending=False).reset_index()
facturacion_por_producto2 = tienda2.groupby('Producto')['facturacion'].sum().sort_values(ascending=False).reset_index()
facturacion_por_producto3 = tienda3.groupby('Producto')['facturacion'].sum().sort_values(ascending=False).reset_index()
facturacion_por_producto4 = tienda4.groupby('Producto')['facturacion'].sum().sort_values(ascending=False).reset_index()


# Combinar los resultados de las facturaciones de cada tienda
facturacion_total = pd.concat([
    facturacion_por_producto.set_index('Producto'),
    facturacion_por_producto2.set_index('Producto'),
    facturacion_por_producto3.set_index('Producto'),
    facturacion_por_producto4.set_index('Producto')
], axis=1, keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
facturacion_total = facturacion_total.reset_index()

# Mostrar la tabla combinada
display(facturacion_total)
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio_promedio_tienda_1 = tienda.groupby('Lugar de Compra')['Costo de envío'].mean().reset_index()
envio_promedio_tienda_2 = tienda2.groupby('Lugar de Compra')['Costo de envío'].mean().reset_index()
envio_promedio_tienda_3 = tienda3.groupby('Lugar de Compra')['Costo de envío'].mean().reset_index()
envio_promedio_tienda_4 = tienda4.groupby('Lugar de Compra')['Costo de envío'].mean().reset_index()

# Concatenar los resultados del envío promedio por tienda
envio_promedio_total = pd.concat([
    envio_promedio_tienda_1.set_index('Lugar de Compra'),
    envio_promedio_tienda_2.set_index('Lugar de Compra'),
    envio_promedio_tienda_3.set_index('Lugar de Compra'),
    envio_promedio_tienda_4.set_index('Lugar de Compra')
], axis=1, keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])

# Restablecer el índice para mostrar el nombre de la tienda como columna normal
envio_promedio_total = envio_promedio_total.reset_index()

# Mostrar la tabla con el envío promedio por tienda
display(envio_promedio_total)
```
